1. What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?
     
   In hypothesis testing, we evaluate a null hypothesis (H₀) against an alternative hypothesis (H₁). Errors can occur because decisions are based on sample data, not the entire population.

*1. Type I Error (False Positive):*

* Occurs when we *reject the null hypothesis even though it is actually true*.
* Example: Concluding a new drug works when it actually does not.
* The probability of making a Type I error is *α (alpha), the *significance level (commonly 0.05).
* *Impact on decision-making:*

  * A Type I error may lead to accepting ineffective or harmful interventions.
  * Stricter α reduces the chance of a Type I error but increases the risk of a Type II error.

*2. Type II Error (False Negative):*

* Occurs when we *fail to reject the null hypothesis when the alternative is actually true*.
* Example: Concluding a new drug does not work when it actually does.
* The probability of making a Type II error is *β (beta)*.
* *Impact on decision-making:*

  * A Type II error may cause us to miss beneficial effects or real differences.
  * Lowering β (increasing test power) requires larger sample sizes.



2. What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

   The *P-value* is the probability of obtaining results *at least as extreme as the observed data, *assuming the null hypothesis (H₀) is true.

   ***How to interpret a P-value:***

* A *small P-value* (usually < 0.05):

  * Suggests the observed data is unlikely under the null hypothesis.
  * Provides evidence *against* H₀, leading us to reject it.

* A *large P-value* (≥ 0.05):

  * Suggests the data is consistent with the null hypothesis.
  * We *fail to reject* H₀ (but we do not prove it true).

***Important notes:***

* The P-value does *not* tell the probability that H₀ is true or false.
* It only measures how surprising the data is *if H₀ were true*.

  

3. Explain the difference between a Z-test and a T-test, including when to use
each.

   - A *Z-test* and a *T-test* are both hypothesis tests used to compare sample data to a population, but they differ mainly in the information available and the sample size.


     ***Z-test:***

     * Used when the *population standard deviation (σ)* is *known*.
     * Suitable for *large samples* (usually *n ≥ 30*).
     * Assumes the sampling distribution follows a *normal distribution*.
     * Test statistic follows the *Z-distribution*.

     ***T-test:***

       * Used when the *population standard deviation is unknown*.
       * Suitable for *small samples* (usually *n < 30*).
       * Uses the *sample standard deviation (s)* as an estimate of σ.
       * Test statistic follows the *t-distribution, which has **heavier tails* and adjusts for smaller sample sizes.



4. What is a confidence interval, and how does the margin of error influence
its width and interpretation?

    - A *confidence interval (CI)* is a range of values, calculated from sample data, that is likely to contain the true *population parameter* (such as the mean or proportion) with a specified level of confidence (e.g., 95%).

     The *margin of error (MOE)* represents how far the estimate (like the sample mean) may be from the true population value.

     *Impact of margin of error:*

     * A ***larger MOE*** produces a **wider confidence interval**, meaning the estimate is less precise but more conservative.
     * A ***smaller MOE*** produces a **narrower confidence interval**, meaning the estimate is *more precise*, but there is less room for variation.

     **Interpretation:**

     A 95% CI means that if we repeated the sampling many times, *95% of those intervals would contain the true population parameter*.


5.  Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

    - **ANOVA** (*Analysis of Variance) is a statistical test used to determine whether there are **significant differences between the means of three or more independent groups*.

     ### *Purpose of ANOVA:*

     * To test whether at least one group mean is different from the others.
     * It extends hypothesis testing beyond two groups, which is what the t-test is limited to.

     ### *How ANOVA Extends Hypothesis Testing:*

    * Instead of performing multiple t-tests (which increases Type I error), ANOVA evaluates *all group means simultaneously* using a single test.
    * It compares the *variance between groups* to the *variance within groups* to determine if group means differ significantly.

     ### *Assumptions of ANOVA:*

      1. *Independence:*
      Observations in each group must be independent of each other.

     2. *Normality:*
     Data in each group should be approximately normally distributed.

    3. *Homogeneity of variance:*
     All groups should have roughly equal variances (tested using Levene’s test).



6. Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.

```python
import numpy as np
from scipy.stats import norm

# Sample dataset
data = np.array([52, 55, 50, 53, 54, 56, 51, 52])
population_mean = 50
population_std = 4  # Known population standard deviation

# Compute sample mean
sample_mean = np.mean(data)

# Z-test formula
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(len(data)))

# Calculate p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")

```

### *Interpretation:*

* The *Z-score* tells how many standard deviations the sample mean is from the population mean.
* The *p-value* shows the probability of observing this sample if the null hypothesis is true.
* If *p < 0.05*, we reject the null hypothesis and conclude the sample mean is significantly different.

8. Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate samples from a non-normal distribution (Exponential)
sample_means = []
num_samples = 1000
sample_size = 50

for i in range(num_samples):
    sample = np.random.exponential(scale=1.0, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot original distribution
data = np.random.exponential(scale=1.0, size=5000)
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Original Non-normal Distribution (Exponential)")

# Plot distribution of sample means
plt.subplot(1,2,2)
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Distribution of Sample Means (CLT)")

plt.tight_layout()
plt.show()
```

### *Explanation:*

* We generate samples from an ***exponential distribution***, which is non-normal.
* As sample size increases, the distribution of sample means becomes approximately normal, demonstrating the **Central Limit Theorem (CLT)**.


9.  Write a Python function to calculate and visualize the confidence interval
for a sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)

    z = norm.ppf((1 + confidence) / 2)
    margin = z * (sample_std / np.sqrt(n))

    lower = sample_mean - margin
    upper = sample_mean + margin

    # Print results
    print("Sample Mean:", sample_mean)
    print(f"{confidence*100}% Confidence Interval: ({lower}, {upper})")

    # Visualization
    plt.axvline(lower, color='red', linestyle='--')
    plt.axvline(upper, color='red', linestyle='--')
    plt.axvline(sample_mean, color='blue')
    plt.hist(data, bins=20, edgecolor='black')
    plt.title("Confidence Interval for Sample Mean")
    plt.show()

    data = np.array([12, 15, 14, 16, 13, 14, 15, 17, 12, 16])
confidence_interval(data)



```

## *Explanation:*

* The function computes the ***95% confidence interval*** using the Z-score.
* It also plots the histogram and highlights:

  * mean (blue line),
  * lower bound, upper bound (red dashed lines).


10. Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example)
observed = np.array([25, 30, 20, 25])

# Expected frequencies
expected = np.array([25, 25, 25, 25])

# Perform Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed distribution differs significantly from expected.")
else:
    print("Fail to reject the null hypothesis: No significant difference between observed and expected distributions.")
```

### *Explanation:*

* The Chi-square goodness-of-fit test checks whether the ***observed frequencies*** match the ***expected frequencies***.
* A low p-value (<0.05) means the distributions are significantly different.

* A high p-value means the data fits the expected distribution well.
